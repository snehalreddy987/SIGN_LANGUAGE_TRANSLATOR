```python
import pandas as pd
import numpy as np

#dataset connection
dataset_path = '/content/sign_mnist_train.csv'
data1 = pd.read_csv(dataset_path)

# Extract labels and pixel values
labels = data1['label'].values
pixels = data1.drop('label', axis=1).values

# Normalize the pixel values to the range [0, 1]
pixels_normalized = pixels / 255.0

# Optionally, reshape the pixel data to match CNN input shape (batch_size, height, width, channels)
pixels_reshaped = pixels_normalized.reshape(-1, 28, 28, 1)

# Print the first few rows of the normalized data
data = pd.DataFrame(pixels_normalized)  # Convert the normalized pixel data back to a DataFrame
data['label'] = labels  # Add the labels back to the DataFrame

data.head()


```

```python


# Check for empty values (NaN) in the normalized pixel values
empty_values = np.isnan(pixels_normalized).sum()
print(f"Total number of empty (NaN) values: {empty_values}")

# Outliers
outliers = np.sum(pixels_normalized > 1)  # Count values greater than 1
print(f"Total number of outliers (values > 1): {outliers}")

outlier_rows = np.where(pixels_normalized > 1)
print(f"Outlier rows and columns: {outlier_rows}")
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Assuming `pixels_normalized` and `labels` are already loaded and preprocessed

# Step 1: Split the data into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(
    pixels_normalized,  # Input features (normalized pixel data)
    labels,             # Output labels
    test_size=0.2,      # 20% data for testing, 80% for training
    random_state=42,    # Set random_state for reproducibility
    stratify=labels     # Ensures that the split has the same proportion of classes as the original dataset
)

# Step 2: Split the training data further into 80% train and 20% validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train,             # Already split training data
    y_train,             # Corresponding labels
    test_size=0.2,       # 20% of the training data for validation
    random_state=42,     # Set random_state for reproducibility
    stratify=y_train     # Ensures the same class distribution in train/val sets
)

# Step 3: Reshape the data to the correct shape for CNN (28, 28, 1)
X_train = X_train.reshape(-1, 28, 28, 1)
X_val = X_val.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Step 4: Define the model (assuming you already have the model defined as 'model')
# Model Definition (Ensure the model is defined only once, not inside the training loop)
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(28, 28, 1)),  # Input layer
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),  # First Conv Layer
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),  # Max Pooling Layer
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),  # Second Conv Layer
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),  # Max Pooling Layer
    tf.keras.layers.Flatten(),  # Flatten the data for Dense layers
    tf.keras.layers.Dense(128, activation='relu'),  # Dense Layer
    tf.keras.layers.Dropout(0.28),  # Dropout Layer
    tf.keras.layers.Dense(26, activation='softmax')  # Output Layer
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Step 5: Train the model for 20 epochs
history = model.fit(
    X_train, y_train,                    # Training data
    epochs=20,                           # Number of epochs
    batch_size=32,                       # Batch size
    validation_data=(X_val, y_val),      # Validation data
    verbose=1                            # Show training progress
)

# Step 6: Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Step 7: Plot the training and validation loss/accuracy
plt.figure(figsize=(12, 6))

# Plot the training and validation accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot the training and validation loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Show the plots
plt.show()

```

```python
import cv2
import numpy as np
```

```python
model = tf.keras.models.load_model('hand_sign_model.h5')

# Define a function to preprocess video frames
def preprocess_frame(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    resized = cv2.resize(gray, (28, 28))  # Resize to 28x28
    normalized = resized / 255.0  # Normalize pixel values
    reshaped = normalized.reshape(1, 28, 28, 1)  # Reshape to match model input
    return reshaped

# Label mapping (index to alphabet)
labels = list("ABCDEFGHIKLMNOPQRSTUVWX")  # Excluding J and Z

# Open webcam (or use a video file by passing its path instead of 0)
video_capture = cv2.VideoCapture(0)  # Use 0 for default webcam

while True:
    ret, frame = video_capture.read()  # Read a frame from the video feed
    if not ret:
        break

    # Predict the hand sign
    processed_frame = preprocess_frame(frame)
    predictions = model.predict(processed_frame)
    predicted_label = np.argmax(predictions)  # Get the class with the highest probability
    predicted_sign = labels[predicted_label]

    # Display the prediction on the frame
    cv2.putText(frame, f"Predicted: {predicted_sign}", (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Show the frame
    cv2.imshow("Hand Sign Detection", frame)

    # Break the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close windows
video_capture.release()
cv2.destroyAllWindows()
```
